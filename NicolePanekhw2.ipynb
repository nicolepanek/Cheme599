# Problem 1

Suppose that a flat area with center at $(x_0, y_0)$ is illuminated by a light source with intensity distribution

$i(x,y) = K e^{-[(x-x_0)^2 + (y-y_0)^2]}$

Assume for simplicity that the reflectance of the area is constant and equal to 1.0, and let K = 255. If the resulting image is digitized with k bits of intensity resolution, and the eye can detect an abrupt change of 8 shades of intensity between adjacent pixels, what value of k will cause false contouring?

Plot the image with k=2, 4, 8, and 16, and 64 intensity levels.

**Hint**: Use the rounding method from the L1 materials to limit the number of possible shades possible.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xi = np.linspace(0,1000,1001).astype(int)
yi = np.linspace(0,1000,1001).astype(int)

x0=500
y0=500
scale = 0.00001 #make close to zero for clear visual

x,y = np.meshgrid(xi,yi)
K = 255
im =K*np.exp(-scale*((x-x0)**2 + (y-y0)**2))

plt.imshow(im, cmap ='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
k = [2,4,8,16,64] #I also did k=32 and k=128 to further show my point

fig, ax = plt.subplots(ncols=5, figsize=(17,5))

for j, axi in enumerate(ax):
  axi.axis('off')
  axi.imshow(256/(k[j]-1)*np.round(im*((k[j]-1)/256), decimals=0), cmap='gray') #divide, round, then rescale to get the resolution 
```

```python
k = [22,32,42] #I am doing k=22, k=32, and k=42 to confirm my answer
fig, ax = plt.subplots(ncols=3, figsize=(10,5))

for j, axi in enumerate(ax):
  axi.axis('off')
  axi.imshow(256/(k[j]-1)*np.round(im*((k[j]-1)/256), decimals=0), cmap='gray')
```

*Since the eye can detect an abrupt change of 8 shades of intensity between adjacent pixels, false contouring occurs when the bits of intensity resolution are 8 or more shades different than eachother. In essense, false contouring occurs when our eyes have better resolution than the image. This is clearly shown in the plots above: when k is less than 32 the image has circular contours visible but the image is a uniform gradient when k>32. (When k=32 there are 8 shades between each allowed intensity - 256/32=8). Above k=32 the bits of intensity resolution in the image is more prescise than our eyes can resolve, so we don't get any false contouring.*

# Problem 2

Consider the image segment below. Let $V$ represent the intensity levels that are set to 1 in a binarized image.

1. Let $V=\{0,1\}$ and compute the lengths of the shortest 4- and 8-path between p and q. If a particular path does not exist between these two points, explain why. Also provide the binarized image used to solve this problem.
2. Repeat for $V = \{1,2\}$. Also provide the binarized image used to solve this problem.

```python
import numpy as np
import matplotlib.pyplot as plt
im = np.array([[3,1,2,1],[2,2,0,2],[1,2,1,1],[1,0,1,2]])
plt.imshow(im, cmap='gray')
plt.text(-0.1, 3, 'p', fontsize=20)
plt.text(3, 0, 'q', fontsize=20)
plt.axis('off')
```

```python
im1=np.zeros((4,4)) 

for a,i in enumerate(im):
  for b,j in enumerate(i):
    if j>1:
      im1[a,b]=0
    else:
      im1[a,b]=1

plt.imshow(im1, cmap='gray') #show the first binarized image
plt.axis('off')
plt.text(-0.1, 3, 'p', fontsize=20)
plt.text(3, 0, 'q', fontsize=20)
print(im1)
```

```python
#location of p and q based on plt.text
#check how to find path besides by inspection

```

*By inspection, the shortest 8-path between p and q is 6 blocks (staying within the white blocks and including p and q). A 4-path does not exist because point q does not have any 4-neighbors with a value of 1.*

```python
im2=np.zeros((4,4)) 

for a,i in enumerate(im):
  for b,j in enumerate(i):
    if j==1:
      im2[a,b]=1
    if j==2:
      im2[a,b]=1
    else:
      im1[a,b]=0

plt.imshow(im2, cmap='gray') #show the second binarized image
plt.axis('off')
plt.text(-0.1, 3, 'p', fontsize=20)
plt.text(3, 0, 'q', fontsize=20)
print(im2)
```

*The shortest 4-path between p and q for the new image is 7 blocks (can be done two ways based on inspection). The shortest 8-path is 5 blocks (can also be done two ways). *
